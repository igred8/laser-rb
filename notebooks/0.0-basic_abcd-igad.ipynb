# UV transport
---

The UV transport has to go from the fourth harmonic generation module (FHG) to the photo cathode. Currently, we are considering a 13 m transport distance from iris to cathode. The iris will be imaged on the cathode without creating a focus on the cathode in order to reduce the risk of "burning" the cathode by delivering a large energy density into a tight focus.

```python
LENS1_POS = 0.0
EX1_DIST = 0.1
DRIFT1_DIST = 10.0
EX2_DIST = 0.1
DRIFT2_DIST = 1.0

namelist = ['lens_01'
           ,'lens_02'
           ,'lens_03'
           ,'lens_04'
           ,'lens_05'
           ]

eletypelist = ['lens','lens','lens','lens','lens']

poslist = [LENS1_POS
          ,LENS1_POS + EX1_DIST
          ,LENS1_POS + EX1_DIST + DRIFT1_DIST
          ,LENS1_POS + EX1_DIST + DRIFT1_DIST + EX2_DIST
          ,LENS1_POS + EX1_DIST + DRIFT1_DIST + EX2_DIST + DRIFT2_DIST
          ]

foclist = [-1.0
          ,1.0
          ,1.0
          ,-1.0
          ,1.50
          ]

elementlist = []
for i,ele in enumerate(namelist):
    elementlist.append(cl.ElementABCD(ele, eletype=eletypelist[i], eleprops={'position':poslist[i], 'focal_len':foclist[i]}))


BLUV = cl.BeamLine()
for i,ele in enumerate(elementlist):
    BLUV.add_element(ele)
```

```python
# make beam
STD_POS = 1E-3
STD_ANG = 200E-6
NMACRO = 50
invec01 = np.stack((np.random.normal(0.0, STD_POS, NMACRO),
                    np.random.normal(0.0, STD_ANG, NMACRO)))
INPOS = -0.01
OUTPOS = np.array([10.0])
outpos01 = np.arange(0.0,15,0.5)
outpos01, outvec01 = BLUV.ray_trace(invec01, INPOS, outpos01)
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(9,6))

ls = dict(linewidth=1,linestyle='-',marker='o',ms=0, mfc='None', alpha=0.91)
ls['color'] = 'dodgerblue'
xvec = outpos01

for i in range(invec01.shape[1]):
    yvec = outvec01[:,0,i]
        
    plt.plot(xvec, yvec,**ls)

plt.style.use('default')
```
